```python
#Create function that cleans the dataframes

def clean_tide_max(df):
    #drops the predicted and premilinary columns as these columns are predicitions not readings
    df = df.drop(columns= ['Predicted (ft)', 'Preliminary (ft)'], axis=1)
    #converts the date column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    #set the index to the date
    df = df.set_index('Date')
    #only keeps rows in the dataframe where we have actual readings
    df = df[df['Verified (ft)'] != '-']
    #converts the verified readings column to a numeric value
    df['Verified (ft) Max'] = pd.to_numeric(df['Verified (ft)'])
    #resample the dataframe to be daily max tide level 
    df = df.resample('D').max()
    #returns the dataframe
    return df
```

```python
def clean_tide_mean(df):
    #drops the predicted and premilinary columns as these columns are predicitions not readings
    df = df.drop(columns= ['Predicted (ft)', 'Preliminary (ft)'], axis=1)
    #converts the date column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    #set the date column to datetime
    df = df.set_index('Date')
    #only keeps rows in the dataframe where we have actual readings
    df = df[df['Verified (ft)'] != '-']
    #converts the verified readings column to a numeric value
    df['Verified (ft) Max'] = pd.to_numeric(df['Verified (ft)'])
    #resample the dataframe to be the daily mean tide reading
    df = df.resample('D').mean()
    #return the dataframe
    return df
```

```python
def clean_baro_max(df):
    #only keep the date column and barometric pressure column
    df = df[['DATE TIME', ' BARO']]
    #rename the columns
    df.rename(columns={'DATE TIME': 'Date', ' BARO': 'baro'}, inplace=True)
    #drop null columns
    df.dropna(inplace=True)
    #set the index to the date column
    df = df.set_index('Date')
    #set the index to datetime
    df.index = pd.to_datetime(df.index)#, format='%Y-%m')
    #converting the barometric pressure column to numeric data types
    df['baro'] = pd.to_numeric(df['baro'])
    #resample the data to be daily max readings
    df = df.resample('D').max()
    #return the dataframes
    return df
```

```python
def clean_baro_mean(df):
    #only keep the date column and barometric pressure column
    df = df[['DATE TIME', ' BARO']]
    #rename the columns
    df.rename(columns={'DATE TIME': 'Date', ' BARO': 'baro'}, inplace=True)
    #drop null columns
    df.dropna(inplace=True)
    #set the index to dateime
    df = df.set_index('Date')
    #change the index to datetime
    df.index = pd.to_datetime(df.index)
    #converting the barometric pressure column to numeric data types
    df['baro'] = pd.to_numeric(df['baro'])
    #resample the data to be daily mean readings
    df = df.resample('D').mean()
    #return the dataframe
    return df
```

```python
#creates fucntion that cleans the unemployment datasets and makes gives them daily unemployment rows
def clean_unemploy(df, noaa_data):
    #set columns as date and unemployment rates
    df.columns = ['Date', 'unemployment_rate']
    #make the date column just the first 7 digits
    df['Date'] = df['Date'].map(lambda x: x[:7])
    #convert the date column to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    #set the index to the Date column
    df = df.set_index('Date')
    #convert the unemployment rate to numeric data types
    df['unemployment_rate'] = pd.to_numeric(df['unemployment_rate'])
    #create list that just has the index from one of the city's tide and baro pressure dataframes
    list_days = noaa_data.index
    #create list that has unemployment amounts for each day of the month it is assigned to
    rate_list = [df.loc[str(i)[:7]]['unemployment_rate'].item() for i in list_days]
    #creates dataframe that has date and unemployment
    unemploy_df = pd.DataFrame(data=rate_list, index=list_days)
    #renames unemployment rate column
    unemploy_df.rename(columns={0: 'rate'}, inplace=True)
    #return the enemploy_df dataframe
    return unemploy_df
```
